```python
import pandas as pd
import sqlalchemy
```

```python
# Import Spotify Data
spotify_data = "spotifyWeeklyTop200Streams.csv"

spotify_df = pd.read_csv(spotify_data)
spotify_df.head(50)
```

```python
# Drop Features Column in Spotify DataFrame
spotify_df = spotify_df.drop(columns = ['Features'])
spotify_df.head()
```

```python
from datetime import timedelta
spotify_df["Week"] = pd.to_datetime(spotify_df["Week"])
spotify_df["onde_day_behind"] = spotify_df["Week"].apply(lambda x: x - timedelta(days=1))
spotify_df.head()
```

```python
# Import the data from Billboard
billboard_data = "billboardHot100_1999-2019.csv"

billboard_df = pd.read_csv(billboard_data)
billboard_df.head()
```

```python
# Clean the Billboard DataFrame
billboard_df = billboard_df.drop(columns = ['Peak.position', 'Weeks.on.chart', 'Date', 'Writing.Credits', 'Lyrics', 'Features'])
billboard_df.head()
```

```python
# Send Spotify DataFrame to a CSV
spotify_df.to_csv('clean_spotify_df.csv')
```
